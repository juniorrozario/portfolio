# **Stock Analysis Project**

In this project, we will analyze FAANG stocks ($META, $AMZN, $AAPL, $NFLX, $GOOG) from 21/03/2022 to 17/03/2023.

Using matploblib and seaborn, we will create visualizations to display comparisons between the stocks. 


```python
# Library imports

import pandas as pd
import numpy as np
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline

import pandas_datareader as pdr
import pandas_datareader.data as web
import yfinance as yf
yf.pdr_override()

from datetime import datetime
```

```python
# Stocks 

stock_list = ['META','AMZN','AAPL','NFLX','GOOG']

# Start and end times for data

end = datetime.now()
start = datetime(end.year - 1, end.month,end.day)

# For loop for gathering yf data as dataframe

for stock in stock_list:
    globals()[stock] = yf.download(stock,start,end)
```

```python
# Historical adj close price of stocks

plt.ylabel('Price Per Share')
plt.title('Adj. Closing Price')

META['Adj Close'].plot()
AMZN['Adj Close'].plot()
AAPL['Adj Close'].plot()
NFLX['Adj Close'].plot()
GOOG['Adj Close'].plot(figsize=(12,5)).legend(stock_list)
```

```python
# Total trading volume from 21/03/2022 to 17/03/2023

plt.title('Total Trading Volume')
plt.ylabel('Trading Volume')

META['Volume'].plot()
AMZN['Volume'].plot()
AAPL['Volume'].plot()
NFLX['Volume'].plot()
GOOG['Volume'].plot(figsize=(12,5)).legend(stock_list)
```

```python
# Moving Average for Adj Closing Price of $META

ma_day = [10,20,50]

for ma in ma_day:

    column_name = 'MA for %s days' %(str(ma))

    META[column_name] = pd.Series(META['Adj Close']).rolling(window=ma).mean()

META[['Adj Close','MA for 10 days','MA for 20 days','MA for 50 days']].plot(subplots=False,figsize=(10,4))

plt.title('Moving Average of Adj. Closing Price of $META')
```

```python
# Closing prices for stock list

closing_prices = web.get_data_yahoo(stock_list,start,end)['Adj Close']

closing_prices
```

```python
# Daily returns for stock_list

returns = closing_prices.pct_change()

returns
```

```python
# Comparison of daily percentage return to check for correlation betweens stocks

returns_plot = sns.PairGrid(returns.dropna())

returns_plot.map_upper(plt.scatter,color='crimson',marker='.')

returns_plot.map_lower(sns.kdeplot,cmap='rocket')

returns_plot.map_diag(plt.hist,bins=30,color='crimson')
```

```python
# Correlation of closing prices

returns_plot = sns.PairGrid(closing_prices.dropna())

returns_plot.map_upper(plt.scatter,color='crimson',marker='.')

returns_plot.map_lower(sns.kdeplot,cmap='rocket')

returns_plot.map_diag(plt.hist,bins=30,color='crimson')
```
